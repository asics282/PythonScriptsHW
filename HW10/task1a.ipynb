Дана функция: f(x) = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)

Определить корни  
Найти интервалы, на которых функция возрастает  
Найти интервалы, на которых функция убывает  
Построить график  
Вычислить вершину  
Определить промежутки, на котором f > 0  
Определить промежутки, на котором f < 0  

```python
from sympy import *
from sympy.abc import x
```

```python
x = Symbol('x')
eq = (sin(x)/2 + cos(x)/2)**2*sin(x)
```

```python
# график функции
plot(eq, (x, -3.5,3.5), line_color='red')
```

```python
# нахождение корней на интервале, где x ∈ [-4, 4]
solve(eq, x)

y=solve(eq)
for i in range(-4, 4):
    print(round (y[i], 2))
```

```python
# нахождение корней уравнения на интервале, где x ∈ [0, 4]
x1=float(y[0])
x2=float(y[1])
x3=float(y[2])
x4=float(y[3])
print(x1, x2, x3, x4)
```

```python
# промежутки возрастания функции
solve(eq > 0)
```

```python
# промежутки убывания функции
solve(eq < 0)
```

```python
# нахождение производной
dif = diff(eq)
print(dif)
# график производной
plot(dif,(x, -3,3), line_color='green')
```

```python
# нахождение экстремумов
solve(dif, x)
y=solve(dif)
array = [] # корни производной
for i in range(0, 6):
    array.append(round (y[i], 2))
print(array)
```
